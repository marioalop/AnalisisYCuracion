# Ejercicio 1: Tablas de Crecimiento

```python
import pandas as pd
import seaborn
import os
import numpy as np
```

```python
BASE_DIR = os.getcwd()
wtageinf = pd.read_csv(BASE_DIR + "/../input/wtageinf.csv")
wtageinf.head()
```

## Graficar las curvas correspondientes a cada percentil para niños y niñas
### a. En la columna Sex, 1 representa a niños y 2 a niñas
### b. Las curvas de los percentiles correspondientes a un mismo sexo deben ir en un mismo
gráfico.
### c. Un gráfico para cada sexo

```python
%matplotlib inline
cols = wtageinf.columns.difference(['Sex', 'Agemos'])
wtageinf[wtageinf.Sex == 1 ].plot(x='Agemos', y=cols, figsize=(20,10))
```

```python
%matplotlib inline
cols = wtageinf.columns.difference(['Sex', 'Agemos'])
wtageinf[wtageinf.Sex == 2 ].plot(x='Agemos', y=cols, figsize=(20,10))
```
